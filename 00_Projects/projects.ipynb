### **OOP Tasks:**
1. **Bank Account Management System:**
   - Create classes for `BankAccount`, `SavingsAccount`, and `CheckingAccount`.
   - Implement methods for deposit, withdraw, and view balance.
   - Use inheritance for specialized account types.
   - Add encapsulation for sensitive information like account balance.

2. **Library Management System:**
   - Create classes for `Library`, `Book`, and `Member`.
   - Implement methods to issue and return books.
   - Use polymorphism for different member types (e.g., `Student`, `Teacher`).

3. **Hospital Management System:**
   - Classes: `Patient`, `Doctor`, `Appointment`.
   - Implement methods for scheduling appointments and viewing patient details.
   - Use abstraction for general methods like scheduling.

4. **E-commerce Platform:**
   - Classes: `Product`, `Cart`, `Order`.
   - Methods to add products to the cart, calculate total, and place an order.
   - Implement getter and setter methods for product details.

5. **School Management System:**
   - Classes: `Student`, `Teacher`, `Course`.
   - Implement methods to enroll students in courses and assign teachers.
   - Use decorators (`@property`) to manage student grades securely.

### **Start Small, Build Big**
Har project ke saath step-by-step OOP concepts ko implement karein, jaise:
- **Abstraction:** Generalized methods.
- **Encapsulation:** Protect sensitive data.
- **Inheritance:** Reusability of code.
- **Polymorphism:** Flexible and dynamic behavior.
